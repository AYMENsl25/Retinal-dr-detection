````python
import shutil
from pathlib import Path

# Your existing dataset
STAGE1 = r'C:\Users\AYMEN\Downloads\data for DR project\Retinal-dr-detection dataset\dataset_stage1'

# Your MAPLES export
MAPLES_EXPORT = r'C:\Users\AYMEN\Downloads\maples_dr_exported'

# Lesion type mapping — MAPLES name → your subfolder name
LESION_MAP = {
    'hemorrhages'       : 'HE',
    'microaneurysms'    : 'MA',
    'exudates'          : 'EX',
    'optic_disc'        : 'OD',
    'neovascularization': 'NV',   # new type!
    'cotton_wool_spots' : 'CW',   # new type!
}

# Make sure new lesion subfolders exist
for code in ['NV', 'CW']:
    Path(f'{STAGE1}/lesion_masks/{code}').mkdir(parents=True, exist_ok=True)
    print(f"✅ Created lesion_masks/{code}/")

# Copy everything
img_count    = 0
vessel_count = 0
lesion_count = 0

for split, tag in [('train', 'train'), ('test', 'test')]:
    split_path = Path(MAPLES_EXPORT) / split

    # ── Copy fundus images ──────────────────────────────────
    fundus_dir = split_path / 'fundus'
    for img_file in sorted(fundus_dir.iterdir()):
        if img_file.suffix.lower() in ('.jpg', '.jpeg', '.png'):
            shutil.copy2(
                img_file,
                Path(STAGE1) / 'images' / f'MAPLES_{tag}_{img_file.name}'
            )
            img_count += 1

    # ── Copy vessel masks ───────────────────────────────────
    vessel_dir = split_path / 'vessels'
    for mask_file in sorted(vessel_dir.iterdir()):
        if mask_file.suffix.lower() in ('.jpg', '.jpeg', '.png'):
            shutil.copy2(
                mask_file,
                Path(STAGE1) / 'vessel_masks' / f'MAPLES_{tag}_{mask_file.name}'
            )
            vessel_count += 1

    # ── Copy lesion masks ───────────────────────────────────
    for maples_name, short_code in LESION_MAP.items():
        lesion_dir = split_path / maples_name
        if not lesion_dir.exists():
            continue
        for mask_file in sorted(lesion_dir.iterdir()):
            if mask_file.suffix.lower() in ('.jpg', '.jpeg', '.png'):
                shutil.copy2(
                    mask_file,
                    Path(STAGE1) / 'lesion_masks' / short_code / f'MAPLES_{tag}_{mask_file.name}'
                )
                lesion_count += 1

print(f"\n✅ Copy complete!")
print(f"   Fundus images copied  : {img_count}")
print(f"   Vessel masks copied   : {vessel_count}")
print(f"   Lesion masks copied   : {lesion_count}")
````

````python
from pathlib import Path

STAGE1 = r'C:\Users\AYMEN\Downloads\data for DR project\Retinal-dr-detection dataset\dataset_stage1'

images       = list(Path(f'{STAGE1}/images').iterdir())
vessel_masks = list(Path(f'{STAGE1}/vessel_masks').iterdir())

sources = {}
for f in images:
    src = f.name.split('_')[0]
    sources[src] = sources.get(src, 0) + 1

print("=" * 50)
print("FINAL DATASET SUMMARY")
print("=" * 50)
print(f"\n  Images by source:")
for src, count in sorted(sources.items()):
    print(f"    {src:<12} : {count}")
print(f"    {'─'*25}")
print(f"    TOTAL      : {len(images)}")
print(f"\n  Vessel masks : {len(vessel_masks)}")
print(f"\n  Lesion masks:")
total = 0
for code in ['MA', 'HE', 'EX', 'SE', 'OD', 'NV', 'CW', 'merged']:
    folder = Path(f'{STAGE1}/lesion_masks/{code}')
    count  = len(list(folder.iterdir())) if folder.exists() else 0
    total += count
    print(f"    {code:<8} : {count}")
print(f"    {'─'*25}")
print(f"    TOTAL    : {total}")
````
